## Interpolación Lineal
---
Nos da como resultado una <font color=gree>recta</font> que se ajusta a dos puntos dados. La forma está dado por
$${\large g(x) = \frac{b - x}{b - a}f(a) + \frac{x - a}{b - a}f(b)}$$



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return x**2

def g(x):
  return (b - x)/(b - a) * f(a) + (x - a)/(b - a) * f(b)
```

```python
a = 4
b = 8
```

```python
plt.figure(figsize = (8,8))

x = np.linspace(0, 10, 100)

plt.plot(x, f(x), label = "La función $f(x)$")
plt.plot(x, g(x), label = "La aproximación $g(x)$")
plt.title("Interpolación lineal", fontsize = 25)

plt.grid()
plt.scatter(a, f(a), c="red", s=60)
plt.scatter(b, f(b), c="red", s=60)

plt.legend()
plt.show()
```
